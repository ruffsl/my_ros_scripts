```python
import pandas as pd
```

```python
svo_ate_error_path = 'svo_ate_error.data'
ebt_ate_error_path = 'ebt_ate_error.data'
omni_ate_error_path= 'omni_ate_error.data'

svo_ate_error  = pd.DataFrame()
ebt_ate_error  = pd.DataFrame()
omni_ate_error = pd.DataFrame()
names = ['stamp', 'x', 'y', 'z']

svo_ate_error = pd.read_csv(svo_ate_error_path, sep=" ", header=None, index_col=2, names=names)
svo_ate_error.index = pd.to_datetime((svo_ate_error.index.values*1e9).astype(int))


ebt_ate_error = pd.read_csv(ebt_ate_error_path, sep=" ", header=None, index_col=0, names=names)
ebt_ate_error.index = pd.to_datetime((ebt_ate_error.index.values*1e9).astype(int))


omni_ate_error = pd.read_csv(omni_ate_error_path, sep=" ", header=None, index_col=0, names=names)
omni_ate_error.index = pd.to_datetime((omni_ate_error.index.values*1e9).astype(int))
```
